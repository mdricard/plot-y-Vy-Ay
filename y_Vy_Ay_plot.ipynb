```python
import numpy as np
import matplotlib.pyplot as plt

fn = "D:/Class Notes/KINE 5350 Applied Biomech/Vertical Jump Ay Vy Y.csv"
data = np.loadtxt(fn, delimiter=",", skiprows=1)
x = data[:,0]
Ry = data[:,1]
Ay = data[:,2]
Vy = data[:,3]
y = data[:,4]

plt.plot(Ry,label="Ry")
plt.ylabel("Vertical Force (N)")
plt.xlabel("Time (ms)")
#plt.legend()
plt.grid(True)
#plt.hlines(y=0,xmin=0,xmax=len(x)-1,colors='y',linestyles='--')
plt.show()
```

```python
plt.plot(Ay,label="Ay")
plt.ylabel("Vertical Acceleration (m/s^2)")
plt.xlabel("Time (ms)")
#plt.legend()
plt.grid(True)
plt.hlines(y=0,xmin=0,xmax=len(x)-1,colors='y',linestyles='--')
plt.show()
```

```python
plt.plot(Vy,'r',label="Vy")
plt.ylabel("Vertical Velocity (m/s)")
plt.xlabel("Time (ms)")
#plt.legend()
plt.grid(True)
plt.hlines(y=0,xmin=0,xmax=len(x)-1,colors='y',linestyles='--')
plt.show()
```

```python
plt.plot(y,'g', label="Y")
plt.ylabel("Vertical Position (m)")
plt.xlabel("Time (ms)")
#plt.legend()
plt.grid(True)
plt.hlines(y=0,xmin=0,xmax=len(x)-1,colors='y',linestyles='--')
plt.show()
```

```python
fig, axs = plt.subplots(3, 1, layout='constrained')
axs[0].plot(y)
axs[0].set_xlabel('Time (ms)')
axs[0].set_ylabel('Position (m)')
plt.hlines(y=0,xmin=0,xmax=len(x)-1,colors='y',linestyles='--')
axs[0].grid(True)

axs[1].plot(Vy, 'r')
#axs[1].legend("Vy")
axs[1].set_xlabel('Time (ms)')
axs[1].set_ylabel('Velocity (m/s)')
plt.hlines(y=0,xmin=0,xmax=len(x)-1,colors='y',linestyles='--')
axs[1].grid(True)

axs[2].plot(Ay, 'g')
#axs[2].legend("Ay")
axs[2].set_xlabel('Time (ms)')
axs[2].set_ylabel('Acceleration (m/s^2)')
plt.hlines(y=0,xmin=0,xmax=len(x)-1,colors='y',linestyles='--')
axs[2].grid(True)

plt.show()
```
